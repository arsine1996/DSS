```python
data = pd.read_excel("Superstore.xlsx",sheet_name=3)
```

```python
data["Ship Mode"].value_counts()/data.shape[0]*100
```

```python
# create a new column, which is equal to 1 when segment is equal to 'Individual', otherwise 0
data["Segment_codes"] = np.where(data.Segment=="Individual",1,0)
```

```python
# define a function, which will return 1, if segment is equal to 'Individual'
def convertor(segment):
    if segment=="Individual":
        return 1
    else:
        return 0
```

```python
data.Segment.apply(convertor).head()
```

```python
# if the numver of unique ellements in a numeric columns is more than 5, build a histogram.
# otherwise build a bar plot

for i in data.select_dtypes(include=["int","float"]).columns.tolist():
    if data[i].nunique()>5:
        data[i].hist(color='lightcoral')
        plt.title(i)
        plt.show()
    else:
        plt.bar(data[i].unique(),data[i].value_counts(),color='navy')
        plt.title(i)
        plt.show()
```
