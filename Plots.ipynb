# Data Visualization with Matplotlib

Matplotlib is the base Python package used for plotting. The intuition behind matplotlib is that you first create the plot object (`plt.figure()`) and then add the plot. If the "frame" of the plot is to be customized (e.g. title, labels, etc.) then specific functions exist in matplotlib for them. Yet, if hte plot itself (the visualization) is requred to customize, then the customization parameters (such as color) should be provided inside the plotting function (such as `plt.hist()`).

`plt.show()` ensures that the plotted visualization will be shown to the user. If multiple plots are generated, then without using `plt.show()` only the last one will be shown.

- [plt.figure(figsize=(8,6))]() - creates the figure istance allowing for customization of size (figsize in inches), figure color etc.
- [plt.title()]() - provides a title for the figure
- [plt.ylabel()]() - provides the label of Y axis
- [plt.xlabel()]() - provides the label of X axis
- [plt.show()]() - shows the figure
- [plt.savefig()]() - saves the figure in a specified format and filename
- [plt.hist()]() - draw a histogram
- [plt.scatter()]() - draws a scatterplot
- [plt.subplot()]() - create subplots (plt.subplot(nrows=1,ncolumns=2,index=1) and plt.subplot(nrows=1,ncolumns=2,index=2) will create 1 row and2 graphs)

```python
plt.figure(figsize=(8,6))
plt.hist(data.price, color="b", bins=15)
plt.title("Car Price Distribution")
plt.ylabel("Frequencies")
plt.xlabel("Price")
plt.show()
```

```python
plt.scatter(data.horsepower,data.price, color="y",marker="+")
plt.title("Relationship between horsepower and price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(14,6))

#plot1
plt.subplot(1,2,1) # number of rows, columns, & graph position
plt.hist(data.price,color="g",bins=15)
plt.title("Car Price Distribution")
plt.ylabel("Frequencies")
plt.xlabel("Price")

#plot2
plt.subplot(1,2,2)
plt.scatter(data.horsepower,data.price,color="y",marker="+")
plt.title("Relationship between horsepower and price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.show()
```

# Data Visualization with Pandas

Pandas is using matplotlib in behind to quickly provide important plotting functionalities to the user. Pandas plots are easy but less customizable. If one wants to customize, sometimes they will have to refer back to matplotlib. Yet, whenever one needs to produce the same plot (e.g. histogram) for several columns, Pandas will be a better choice (otherwise one needs to use matplotlib inside a for loop).



```python
#let's see the columns in the dataframe
print(data.columns.tolist())
```

```python
#choose some columns for plotting
columns_to_plot = ['length', 'width', 'height', 'price']
```

```python
#plot histogram for one of them using matplotlib
plt.hist(data['price'])
plt.show()
```

```python
#plot histogram for all of them using matplotlib and for loop
for i in columns_to_plot:
    plt.hist(data[i])
    plt.title(i)
    plt.show()
```

```python
#plot histogram for all numeric columns in dataframe just using pandas
data.hist(figsize=(20,20)) #histogram of all columns
plt.show()
```

# Data Visualization with Seabron

Seaborn is a user friendly backage developed on top of matplotlib providing quick functionality for developing key analytical visualizations. It is usually easier than atplotlib to use and creates quite user friendly plots. Yet, specific custimizations may still require usage of matplotlib.

- [sns.distplot()](https://seaborn.pydata.org/generated/seaborn.distplot.html) -provides the univariate distribution plot (histogram and and a line plot) for a given variable
- [sns.heatmap()](https://seaborn.pydata.org/generated/seaborn.heatmap.html) - creates a heatmap out of the input matrix

```python
sns.distplot(data.price)
plt.show()
```

```python
#create heatmap of correlation matrix
plt.figure(figsize=(20,20))
#annot=True will also show the correlation values
sns.heatmap(data.corr(), annot=True,cmap='PuBu')
plt.show()
```
