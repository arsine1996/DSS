### ForLoop & ListComprehension Examples

```python
my_even=[]
for i in range(10,30,2):
    my_even.append(i)
print(my_even)
```

```python
my_even=[]
my_odd=[]
for i in range(10,30):
    if i%2==0:
        my_even.append(i)
    else:
        my_odd.append(i)
print('my_even',my_even)
print('my_odd',my_odd)
```

```python
my_even=[i for i in range(10,30) if i%2==0]
my_even
```

```python
{i:i**2 for i in range(10,30) if i%2==0}
```

```python
my_odd=[]
i=10
while i<30:
    if i%2!=0:
        my_odd.append(i)
    i=i+1
print(my_odd)
```
